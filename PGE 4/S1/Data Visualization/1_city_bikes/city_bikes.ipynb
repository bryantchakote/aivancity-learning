3. Write a function which takes as parameters:
    - *n*, a number of intervals
    - *city*
    
    and returns the distribution of stations (count) which fall into the following filling rates interval: $[[0, \frac{1}{n}], (\frac{1}{n}, \frac{2}{n}], \dots, (\frac{n-2}{n}, \frac{n-1}{n}], (\frac{n-1}{n}, 1]]$. In other words, the returned value should be a list of size *n*.
    
The function should ignore non-functioning stations (stations where there is no bike available and no empty slots). 

```python
import numpy as np
import matplotlib.pyplot as plt

def get_fill_distribution(n, city):
    stations = get_stations_info(city)
    data = [station["free_bikes"] / (station["free_bikes"] + station["empty_slots"]) for station in stations if station["free_bikes"] + station["empty_slots"] != 0]
    ranges = np.linspace(0, 1, n+1)
    
    counts = [sum(list(map(lambda x: ranges[i] < x <= ranges[i+1], data))) for i in range(n+1) if i != n]
    counts[0] += sum(list(map(lambda x: x == 0, data)))
    
    bounds = ["]" + str(round(ranges[i], 2)) + ", " + str(round(ranges[i+1], 2)) + "]" for i in range(1, n+1) if i != n]
    bounds.insert(0, "[" + str(round(ranges[0], 2)) + ", " + str(round(ranges[1], 2)) + "]")
    plt.subplots(figsize=(12, 8))
    plt.bar(bounds, counts)
    
    return counts
```
